```python
# Create a list of length 10 with multiple data types
my_list = [
    13,                 # int
    3.1415,               # float
    "Manthan",           # string
    True,               # boolean
    [1, 2, 3],          # list
    (4, 5),             # tuple
    {"a": 1},           # dictionary
    {1, 2, 3},          # set
    None,               # NoneType
    2 + 3j              # complex
]

for item in my_list:
    print(item, "->", type(item))
```
