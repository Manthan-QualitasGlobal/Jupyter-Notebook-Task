```python
# MUTABLE DATA TYPE
print("Mutable Data Type Example (List):")
my_list = [1, 2, 3]
print("Original list:", my_list)

my_list[0] = 10
print("After changing first element:", my_list)

my_list.append(4)
print("After appending 4:", my_list)

print("\n") 

# IMMUTABLE DATA TYPE
print("Immutable Data Type Example (String):")
my_str = "Hello"
print("Original string:", my_str)

my_str = "h" + my_str[1:]
print("Modified string:", my_str)
```
